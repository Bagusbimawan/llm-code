### Import Library

```python
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    T5Config,
    T5ForConditionalGeneration,
    Trainer,
    TrainingArguments,
)
import torch
import multiprocessing
```

```python
# --- 2. Load dataset dari Hugging Face (harus login dulu)
# Dataset ini adalah kumpulan soal reasoning kode Python + jawabannya
ds = load_dataset("nvidia/OpenCodeReasoning", "split_0")
```

```python
train_data = ds["split_0"].remove_columns([col for col in ds["split_0"].column_names if col not in ['input', 'output']])
```

```python
print(train_data)
```

## Tokenizer

```python
tokenizer = AutoTokenizer.from_pretrained("Salesforce/codet5-small")
```

```python
num_proc = multiprocessing.cpu_count()
```

```python
def preprocess_function(example):
    model_input = tokenizer(
        example["input"],
        max_length=512,
        padding="max_length",
        truncation=True
    )
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            example["output"],
            max_length=128,
            padding="max_length",
            truncation=True
        )
    model_input["labels"] = labels["input_ids"]
    return model_input

tokenized_dataset = train_data.map(
    preprocess_function,
    batched=True,
    remove_columns=train_data.column_names,
    num_proc=num_proc
)

tokenized_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
```
